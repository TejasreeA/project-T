```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv('/content/Salary_Data (1).csv')

print (dataset)
```

```python
x= dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print(x)
print(y)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.title('area va.price')
plt.scatter(x,y,color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train ,y_test=train_test_split(x,y,test_size=0.3)
print(x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
print(y_train)
```

```python
accuracy=model.score(x_test,y_test)
print(accuracy)
```

```python
plt.scatter(x_train,y_train,color='blue',marker='*')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.title('yearsexperience vs salary (taining data)')
plt.plot(x_train,model.predict(x_train),color='red')
plt.show()

plt.scatter(x_test,y_test,color='blue',marker='+')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.title('yearsexperience vs salary ( test dataset)')
plt.plot(x_test,model.predict(x_test),color='red')
plt.show()
```
